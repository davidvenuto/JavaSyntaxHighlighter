```python
import pandas as pd
from pygments.lexers import JavaLexer
from pygments import lex
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('parsed_code.csv', delimiter=';')
df
```

```python
label_encoder = LabelEncoder()
type_integers = label_encoder.fit_transform(df['Type'])
type_categorical = to_categorical(type_integers)
X_train, X_test, y_train, y_test = train_test_split(type_integers.reshape(-1, 1), type_categorical, test_size=0.2, random_state=42)
```

```python
model = Sequential([
    Dense(64, activation='relu', input_shape=(1,)),
    Dense(32, activation='relu'),
    Dense(type_categorical.shape[1], activation='softmax')  # Output layer with one neuron per type
])

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(X_train, y_train, epochs=50, validation_split=0.2)
```
